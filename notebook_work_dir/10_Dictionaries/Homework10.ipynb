Ключевой идеей шфра Цезаря является замещение каждого символа другим символом с заданним смещением в алфавите.
Например, если мы захотим создать шифр со смещением 3, нам понадобится следующая карта символов:

Исходные:   ABCDEFGHIJKLMNOPQRSTUVWXYZ
Шифрованные:  DEFGHIJKLMNOPQRSTUVWXYZABC

Чтобы создать подобный шифр, нужно иметь представление о том, как представлены символы в компьютере.

Например, ASCII код символа ’A’ равен 65 и ASCII код символа ’a’ равен 97. 
Для конвертации ASCII кодов и символом в Python можно исползовать следующий код:
</pre>

```python
letter = 'a'
# конвертируем символ в ascii код
ascii_code = ord(letter)
# конвертируем ascii код в символ
letter_res = chr(ascii_code)
print ascii_code, letter_res
```
